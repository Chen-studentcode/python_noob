##Comprehension 大略格式: 變數=[執行(結果) 迴圈 判斷] 

###一般迴圈-01

```
data=[23,77,88,99,56,71,23,24,6,8]
odds=[]

for num in data:
    if num%2:
        odds.append(num)
 
```

###Comprehension寫法-01

```
data=[23,77,88,99,56,71,23,24,6,8]
odds=[]

odds_c=[num for num in data if num%2]
```

###一般迴圈-02

```
text='We are the champions my friends'
data=text.split(' ')
for i in data: 
    print(i)
```

###Comprehension寫法-02

```
data=[print(i) for i in 'We are the champions my friends '.split(' ')]
```
